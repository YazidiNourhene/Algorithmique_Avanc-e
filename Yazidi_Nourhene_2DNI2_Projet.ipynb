```python
#Check row for a queen, if there is a queen returns true, else returns false.
def row_check(n, r, c, board):
    for i in range (0,n):
        if board [r][i] == 'y':
            return True
    return False
```

```python
#Check column for a queen, if there is a queen returns true, else returns false.
def column_check(n, r, c, board):
    for x in range (0,n):
        if board [x][c] == 'y':
            return True
    return False
```

```python
#Check diagonal for a queen, if there is a queen returns true, else returns false.
def diagonal_check(n, r, c, board):
    rd = r
    cd = c
    while rd>0 and cd>0:
        rd = rd-1
        cd = cd-1
        if board [rd][cd] == 'y':
            return True
        break
    rd = r 
    cd = c
    while rd>0 and cd < (n-1):
        rd = rd-1
        cd = cd+1
        if board [rd][cd] == 'y':
            return True
        break
    return False
```

```python
#Determines whether placing a queen in a spot is safe given the game board
def placer_reine_possible(n, r, c, board):
    if not row_check(n, r, c, board) and not column_check(n, r, c, board) and not diagonal_check(n, r, c, board):
        return True
    else:
        return False
```

```python
#Placing all queens using recursion function
def placer_toutes_reines(n, r, board):
    c = 0
    while c < n:
        if placer_reine_possible(n, r, c, board):
            board[r][c] = 'y'
            if r == n-1:
                return True
            if not (placer_toutes_reines(n, r+1, board)):
                board[r][c] = 0
                c = c+1
            else:
                return True
        else:
            c = c + 1
    return False
```

```python
#Tester l'algorithme
def Test(n):
    board = [[0 for i in range (n)] for x in range (n)]
    if placer_toutes_reines(n, 0, board):
        for r in range (0,n):
            for c in range (0,n):
                if board[r][c] == 0:
                    board [r][c] = '1'
        for i in range (0,n):
            print(" ".join(board[i]))
    else:
        print('no solution')
```
