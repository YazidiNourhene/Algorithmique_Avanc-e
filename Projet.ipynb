```python
def row_check(n, r, c, board):
    for i in range (0,n):
        if board [r][i] == 'Q':
            return True
    return False
```

```python
def column_check(n, r, c, board):
    for x in range (0,n):
        if board [x][c] == 'Q':
            return True
    return False
```

```python
def diagonal_check(n, r, c, board):
    rd = r
    cd = c
    while rd>0 and cd>0:
        rd = rd-1
        cd = cd-1
        if board [rd][cd] == 'Q':
            return True
        break
    rd = r 
    cd = c
    while rd>0 and cd < (n-1):
        rd = rd-1
        cd = cd+1
        if board [rd][cd] == 'Q':
            return True
        break
    return False
```

```python
def move_safe(n, r, c, board):
    if not row_check(n, r, c, board) and not column_check(n, r, c, board) and not diagonal_check(n, r, c, board):
        return True
    else:
        return False
```

```python
def solve_n_queens(n, r, board):
    c = 0
    while c < n:
        if move_safe(n, r, c, board):
            board[r][c] = 'Q'
            if r == n-1:
                return True
            if not (solve_n_queens(n, r+1, board)):
                board[r][c] = 0
                c = c+1
            else:
                return True
        else:
            c = c + 1
    return False
```

```python
def start(n):
    board = [[0 for i in range (n)] for x in range (n)]
    if solve_n_queens(n, 0, board):
        for r in range (0,n):
            for c in range (0,n):
                if board[r][c] == 0:
                    board [r][c] = '_'
        for i in range (0,n):
            print(" ".join(board[i]))
    else:
        print('no solution')
```
